```python
import pandas as pd
import openpyxl
import matplotlib.pyplot as plt
```

```python
usage_data = pd.read_excel("input data\\Safety Stock Çalışma.xlsx",sheet_name="USAGE")
lead_time_data = pd.read_excel("input data\\Safety Stock Çalışma.xlsx",sheet_name="DATA")
```

```python
df = usage_data.transpose() # After transpose: Rows: Time, Columns: Raw Material Data
df = df.rename(columns=df.iloc[0]).drop(df.index[0])
df = df.dropna()

df[df.columns[:]] = df[df.columns[:]].apply(pd.to_numeric) # columns to numeric

clean_df = df.loc[:, (df != 0).any(axis=0)] #Remove raw materials with all 0 values (Kullanılmayalan RM ler)
```

```python
clean_df.index = pd.to_datetime(clean_df.index ) #Datetime the index
clean_df = clean_df.loc[(df!=0).any(axis=1)] # Remove holiday data, where all rm usage is 0 
RM_list = clean_df.columns.copy() # RM names, all 0 RMs removed, (118 Raw materials)
clean_df #330 days 118 RMs
```

```python
clean_df['month'] = clean_df.index.month
clean_df['week'] = clean_df.index.week # Get week of days in data
clean_df['year'] = clean_df.index.year # Get year of days in data

#grouped_by_week_averages = clean_df.groupby(['week','year']).mean() #Weekly averages 
grouped_by_month_sum = clean_df.groupby(['month','year']).sum() # Monthly usage
grouped_by_month_sum
```

```python
grouped_by_month_sum.reset_index()
grouped_by_month_sum = grouped_by_month_sum.sort_values(['year','month'],ascending=[True,True])
grouped_by_month_sum.drop('week',inplace = True,axis=1)
grouped_by_month_sum
```

```python
grouped_by_month_std = clean_df.groupby(['month','year']).std() #Weekly standart deviations
z_score = 1.96
    
# Calculate monthly safety stock for each Raw Material
for rm_id in list(RM_list):
    lead_time = lead_time_data.loc[lead_time_data['MATERIAL NUMBER'] == rm_id, 'TRANSIT TIME'].iloc[0] 
    rm_data = grouped_by_month_std[rm_id]
    safety_stock_data =  rm_data * lead_time * z_score 

    if len(grouped_by_month_safety_stock) != 0:
        grouped_by_month_safety_stock[rm_id] = safety_stock_data
    else:
        grouped_by_month_safety_stock = safety_stock_data    
        grouped_by_month_safety_stock = pd.DataFrame(grouped_by_month_safety_stock)
```
